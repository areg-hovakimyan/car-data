```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import string
```

```python
df = pd.read_csv('volkswagen_gti_2017-2020.csv')
```

```python
df.head()
```

```python
df.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
df.head()
```

```python
df['avg_price'] = df['avg_price'].str.replace('[$ ]' ,'')
df['avg_price'] = df['avg_price'].astype(int)
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 8))
plot = sns.histplot(data = df, y = 'documents')
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
sns.boxplot(df, x ='price')
```

```python
df.head()
```

```python

def clean_text(text):
    text = ''.join([char for char in text if char not in string.ascii_letters + string.punctuation])
    return text
df['mileage'] = df['mileage'].apply(clean_text)
df['mileage'] = df['mileage'].astype(int)
```
